# Little bit of everything
- Matplotlib styling
- Importing an Excel Book
- Grouping
- Plotting arguments

```python
import pandas as pd
import matplotlib
%matplotlib inline
matplotlib.style.use('fivethirtyeight')
xl = pd.ExcelFile('Datasets/ANTIMONY.XLS')
df = xl.parse(xl.sheet_names[0])
for title, grp in df.groupby('Method'):
        grp.plot.hist(subplots=True, layout=(1,2), sharex=False, sharey=True, title=title)
```
