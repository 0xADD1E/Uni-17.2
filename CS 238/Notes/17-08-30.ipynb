```python
from IPython.display import display, Markdown
import math
sign = lambda x: math.copysign(1, x)
ins = [['Bit Code', 'Unsigned', 'Signed Magnitude', '2s Compliment'], ['---', '---', '---', '---']]
for i in range(0,256):
    a = bin(i).replace('0b','')
    b = '0'*(8-len(a))
    ins.append([b+a, i])
for i in ins[2:]:
    bits = i[0]
    x = 0
    y = 1
    z = 0
    if bits[0] == '1':
        x = 128
        y = -1
        z = -256
    i.append(y*i[1]+x)
    t = bits.replace('0', 'x')
    t = t.replace('1', '0')
    t = t.replace('x', '1')
    t = int(t, 2)
    u = t * sign(t)**2
    u = u % 128
    u = u * sign(t)
    i.append(int(u))
md = '\n'.join(['|{}|{}|{}|{}|'.format(str(x[0]), str(x[1]), str(x[2]), str(x[3])) for x in ins])
display(Markdown(md))
```
